```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv("Zomato data.csv")

```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe["listed_in(type)"])
plt.xlabel("type of resturant")
```

```python
grouped_data=dataframe.groupby("listed_in(type)").votes.sum()
result=pd.DataFrame({"vds":grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of resturant",c="black",size=20)
plt.ylabel("votes",c="black",size=20)
```

```python
plt.hist(dataframe["rate"],bins=5)
plt.title("rating distrubrution")
plt.show()
```

```python
sns.countplot(x=dataframe["approx_cost(for two people)"])
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="online_order",y="rate",data=dataframe)
```

```python
df=dataframe.groupby(["listed_in(type)","online_order"]).size().unstack(fill_value=0)
sns.heatmap(df,annot=True,cmap="YlGnBu")
plt.title("Heatmap")
plt.xlabel("Online order")
plt.ylabel("listed in(type)")
plt.show()
```

```python
online_order_counts = dataframe['online_order'].value_counts()
plt.pie(online_order_counts, labels=online_order_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title("Online Order Availability")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=dataframe["approx_cost(for two people)"], y=dataframe["votes"],hue=dataframe["listed_in(type)"])
plt.title("Cost vs Votes")
plt.xlabel("Cost (for two people)")
plt.ylabel("Votes")
plt.show()
```
